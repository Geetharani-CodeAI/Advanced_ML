```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
```

```python
def scaler_split(x, y):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 0)
    sc = StandardScaler()
    x_train = sc.fit_transform(x_train)
    x_test = sc.transform(x_test)
    return x_train, x_test, y_train, y_test
```

```python
# Applying PCA
def pca(x_train, x_test, n):
    pca = PCA(n_components = n)
    x_train_pca = pca.fit_transform(x_train)
    x_test_pca = pca.transform(x_test)
    explained_variance = pca.explained_variance_ratio_   # Explained_variance- measure of percentage of total variance in a dataset for each component 
    return x_train_pca, x_test_pca, explained_variance
```

```python
def get_r_score(regressor, x_test_pca, y_test):
    y_pred = regressor.predict(x_test_pca)
    r_score = r2_score(y_test, y_pred)
    return r_score
```

```python
def linear(x_train_pca, y_train, x_test_pca, y_test):
    regressor = LinearRegression()
    regressor.fit(x_train_pca, y_train)
    r_score = get_r_score(regressor, x_test_pca, y_test)
    return regressor, r_score, x_test_pca, y_test
```

```python
def svml(x_train_pca, y_train, x_test_pca, y_test):
    regressor = SVR(kernel = 'linear')
    regressor.fit(x_train_pca, y_train)
    r_score = get_r_score(regressor, x_test_pca, y_test)
    return regressor, r_score, x_test_pca, y_test
```

```python
def svmnl(x_train_pca, y_train, x_test_pca, y_test):
    regressor = SVR(kernel = 'rbf')
    regressor.fit(x_train_pca, y_train)
    r_score = get_r_score(regressor, x_test_pca, y_test)
    return regressor, r_score, x_test_pca, y_test
```

```python
def decision(x_train_pca, y_train, x_test_pca, y_test):
    regressor = DecisionTreeRegressor(criterion = 'squared_error', random_state = 0)
    regressor.fit(x_train_pca, y_train)
    r_score = get_r_score(regressor, x_test_pca, y_test)
    return regressor, r_score, x_test_pca, y_test
```

```python
def random(x_train_pca, y_train, x_test_pca, y_test):
    regressor = RandomForestRegressor(random_state = 0)
    regressor.fit(x_train_pca, y_train)
    r_score = get_r_score(regressor, x_test_pca, y_test)
    return regressor, r_score, x_test_pca, y_test
```

```python
def pca_regression(acclin, accsvml, accsvmnl, accdeci, accrand):
    dataframe = pd.DataFrame( index = ['PCA'], columns = ['Linear', 'SVML', 'SVMNL', 'Decision', 'Random'])
    for number, item in enumerate(dataframe.index):
        dataframe['Linear'][item] = acclin[number]
        dataframe['SVML'][item] = accsvml[number]
        dataframe['SVMNL'][item] = accsvmnl[number]
        dataframe['Decision'][item] = accdeci[number]
        dataframe['Random'][item] = accrand[number]
    return dataframe        
```

```python
dataset = pd.read_csv('prep.csv', index_col = None)  
dataset
```

```python
df1 = dataset
df1 = pd.get_dummies(df1, dtype = int, drop_first = True)
```

```python
df1
```

```python
x = df1.iloc[:,0:28].values
y = df1.iloc[:,27].values
```

```python
x.shape
```

```python
x_train, x_test, y_train, y_test = scaler_split(x, y)
```

```python
acclin = []
accsvml = []
accsvmnl = []
accknn = []
accnaive = []
accdeci = []
accrand = []
n_components_list = []
```

```python
for n in range(1, 4):  # for PCA with 1 to 3 components
    x_train_pca, x_test_pca, explained_variance = pca(x_train, x_test, n)
 
    regressor, r_score, x_test_pca, y_test = linear(x_train_pca, y_train, x_test_pca, y_test)
    acclin.append(r_score)
    
    regressor, r_score, x_test_pca, y_test = svml(x_train_pca, y_train, x_test_pca, y_test)
    accsvml.append(r_score)
    
    regressor, r_score, x_test_pca, y_test = svmnl(x_train_pca, y_train, x_test_pca, y_test)
    accsvmnl.append(r_score)
    
    regressor, r_score, x_test_pca, y_test = decision(x_train_pca, y_train, x_test_pca, y_test)
    accdeci.append(r_score)
    
    regressor, r_score, x_test_pca, y_test = random(x_train_pca, y_train, x_test_pca, y_test)
    accrand.append(r_score)
    
    n_components_list.append(n)
  
```

```python
results_df = pd.DataFrame({
    'PCA': n_components_list,
    'Linear': acclin,
    'SVML': accsvml,
    'SVMNL': accsvmnl,
    'Decision': accdeci,
    'Random': accrand
})


results_df.set_index('PCA', inplace=True)

# Print the final result table
print(results_df)
```
